```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([1,2,3,4,5])
print(arr)
print(arr[0:3])
print(arr[2])
print(arr[1] , "",  arr[4])
print(arr[1] + arr[3])
```

```python
country = ["Pakistan", "England", "Australia", "South Africa"]
score = [200,210,223,160]

plt.title("Bar Chart")
plt.xlabel("Country")
plt.ylabel("Score")
plt.bar(country, score)
plt.show()
```

```python
plt.title("Line Chart")
plt.xlabel("Country")
plt.ylabel("Score")
plt.grid(True)
plt.plot(country, score)
plt.show()
```

```python
plt.title("Scatter Plot Chart")
plt.xlabel("Country")
plt.ylabel("Score")
plt.grid(True)
plt.scatter(country, score)
plt.show()
```

```python
plt.title("Histogram Chart")
plt.xlabel("Country")
plt.ylabel("Score")
plt.hist(country)
plt.show()

print("\n=================================================================================================\n")


plt.title("Histogram Chart")
plt.xlabel("Country")
plt.ylabel("Score")
plt.hist(score)
plt.show()
```

```python
plt.title("Pie Chart")
mylabels = ["Pakistan", "England", "Australia", "South Africa"]
plt.pie(score, labels= mylabels)
plt.show()
```
